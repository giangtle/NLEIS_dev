```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("NLEIS.txt", skiprows=1, sep='\t')
```

```python
df.columns
```

## First harmonic response (EIS response)

```python
fig, ax = plt.subplots(figsize=(7,4))
ax.plot(df["V11 Re"], -df["V11 Im"])
ax.scatter(df["V11 Re"][::10], -df["V11 Im"][::10], marker="s")
ax.scatter(0, 0, c="r", marker="o")
ax.axis('equal')
ax.set_xticks(np.arange(0,280,40))
ax.set_yticks(np.arange(-40,120,40))
ax.set_xlabel(r"$Z_{11}' [\Omega]$", fontsize=12)
ax.set_ylabel(r"$-Z_{11}'' [\Omega]$", fontsize=12)
plt.grid()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(7,4))
ax.plot(df["V22 Re"], df["V22 Im"])
ax.scatter(df["V22 Re"][::10], df["V22 Im"][::10], marker="s")
ax.scatter(0, 0, c="r", marker="o")
ax.axis('equal')
ax.set_xlabel(r"$Z_{22}' [\Omega/A]$", fontsize=12)
ax.set_ylabel(r"$Z_{22}'' [\Omega/A]$", fontsize=12)
plt.grid()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(7,4))
ax.plot(df["V33 Re"], df["V33 Im"])
ax.scatter(df["V33 Re"][::10], df["V33 Im"][::10], marker="s")
ax.scatter(0, 0, c="r", marker="o")
ax.axis('equal')
ax.set_xlabel(r"$Z_{33}' [\Omega/A^2]$", fontsize=12)
ax.set_ylabel(r"$Z_{33}'' [\Omega/A^2]$", fontsize=12)
ax.set_yticks(np.arange(-1.0E6,3.0E6,1.0E6))
plt.ticklabel_format(style='sci', axis='both', scilimits=(0,0))
plt.grid()
plt.show()
```
